```python
#Seaborn :-


import seaborn as sns

sns.set_theme()

tips =  sns.load_dataset("tips")
print(tips.head())
```

```python
sns.relplot(data= tips ,x = "total_bill" ,y= "tip" ,
            col = "time" ,hue = "smoker" ,size = "size")

```

```python
dots = sns.load_dataset("dots")

print(dots.head())

```

```python
sns.relplot(data = dots , x = "time" , y = "firing_rate" , kind = "line",
           hue = "choice" , col = "align", size = "coherence" , style = "choice")


```

```python
fmri = sns.load_dataset("fmri")
print(fmri.head())
```

```python
sns.relplot(data = fmri , x = "timepoint" , y = "signal" , col = "region" ,kind = "line",
           hue = "event" ,  style= "event")

```

```python
sns.lmplot(data = fmri ,x = "timepoint" , y = "signal" ,col = "region")

```

```python
sns.lmplot(data = tips , x = "total_bill" , y = "tip" , col = "time")

```

```python
#kde :- kernal estimation density
sns.displot(data = tips ,x = "total_bill"  , col = "time", kde= True)

```

```python
# ecdf :-  estimation of cumulative distribution function

sns.displot(data = tips , kind = "ecdf" , col = "time" ,x = "total_bill", hue = "smoker" ,rug = True )
```
